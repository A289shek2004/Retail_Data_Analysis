```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

products = pd.read_csv("../data/raw/products.csv")
products.head()
```

```python
cities = ["Mumbai", "Delhi", "Bengaluru", "Pune", "Hyderabad", "Chennai", "Ahmedabad"]

customers = []

for i in range(1, 501):   # 500 customers
    customers.append({
        "customer_id": f"C{i:04d}",
        "city": random.choice(cities),
        "gender": random.choice(["M", "F"]),
        "age_group": random.choice(["18-25", "26-35", "36-45"])
    })

customers_df = pd.DataFrame(customers)
customers_df.to_csv("../data/raw/customers.csv", index=False)
```

```python
orders = []

start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)

order_id = 1

for _ in range(3000):   # 3000 orders
    customer = customers_df.sample(1).iloc[0]
    product = products.sample(1).iloc[0]

    order_date = start_date + timedelta(
        days=random.randint(0, (end_date - start_date).days)
    )

    quantity = random.choice([1, 1, 1, 2])  # mostly 1
    price = product["price"]
    revenue = quantity * price

    orders.append({
        "order_id": f"O{order_id:05d}",
        "customer_id": customer["customer_id"],
        "product_id": product["product_id"],
        "order_date": order_date.date(),
        "quantity": quantity,
        "price": price,
        "revenue": revenue,
        "city": customer["city"]
    })

    order_id += 1

orders_df = pd.DataFrame(orders)
orders_df.to_csv("../data/raw/orders.csv", index=False)
```

```python
return_reasons = {
    "Bottle": ["Leakage", "Damaged"],
    "Lunch Box": ["Damaged", "Missing item"],
    "Accessories": ["Quality issue"]
}

returns = []

for _, row in orders_df.sample(frac=0.1).iterrows():  # ~10% returns
    product_type = products.loc[
        products["product_id"] == row["product_id"], "product_type"
    ].values[0]

    returns.append({
        "order_id": row["order_id"],
        "return_reason": random.choice(return_reasons.get(product_type, ["Other"]))
    })

returns_df = pd.DataFrame(returns)
returns_df.to_csv("../data/raw/returns.csv", index=False)
```
